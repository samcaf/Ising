# Ising Type Model Tests: Mixed Field Ising Model

Imports:

```python
# Basic imports
import numpy as np
import matplotlib.pyplot as plt

# Local imports
from ising.models.models import ising_model 
# import ising.utils.calculation_utils as cu
```

## Exact Diagonalization Example:

```python
# Base parameters:
L = 14
J = 1

# Critical MFIM parameters:
params = {'name': 'MFIM',
          'L': 10,
          'S': 1/2,
          'bc': 'pbc',
          'J': 1,
          'hx': .55,
          'hz': 0}
```

#### Symmetry, Hamiltonian, and Eigensystem Generation
Unless otherwise specified the model generates symmetry projectors, a Hamiltonian, and the associated eigensystem organized into symmetrysectors, upon initialization.

```python
# Generating a large MFIM model for exact diagonalization:
mfim_exact = ising_model(**params)
```

#### Plotting
Plotting features of the model using built-in model functions.

```python
# Plotting the eigenvalue distribution (against GOE by default):
mfim_exact.plot_eval_dist()

# Plotting the level spacing distribution:
print(mfim_exact.projectors.keys())
for sector in mfim_exact.projectors.keys():
    mfim_exact.plot_levelspace_dist(sector)
```

## Ground State Examples:
If we work in the ground state, we may efficiently deal with systems of larger size.

```python
# Generate a model of larger size:
params['L'] = 18
mfim_large = ising_model(fast_start=False, **params)
# Find the ground state of this model:
mfim_large.get_H()
mfim_large.gen_eigensys(use_symms=False, k=1, which='SA')
```

```python
# Ground state and energy:
E_gs = mfim_large.eigensys['evals'][0]
print("Ground state energy:", E_gs)
print("Ground state in computational basis:", mfim_large.eigensys['evecs'][0])
```

```python
# Entanglement in the ground state
print("Ground state entanglement:")
mfim_large.plot_entanglement(eval=E_gs)
```
