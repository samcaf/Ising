# Entanglement Entropy for Critical Models

Imports:

```python
# Basic imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import eigsh

# Local imports
import ising.models.base as base
import ising.utils.calculation_utils as cu
```

# Base Ising-type models

## Critical Mixed Field Ising Model

See http://edu.itp.phys.ethz.ch/fs13/cft/SM_Molignini.pdf

```python
# Base parameters:
L = 16
J = 1
cut_lens = np.array(range(L-1))+1

# Critical MFIM parameters:
mfim_critical_params = {'hx': 0,
                        'hz': J*.55}
# Notice that hz multiplies S_z = sigma_z/2, and J multiplies S_x S_x
```

```python
# Generating a large MFIM model Hamiltonian:
H_mfim = base.gen_mixedfieldising(L=L, J=J, bc='pbc', **mfim_critical_params)
```

```python
# Find the ground state of this model:
_, ground_state = eigsh(H_mfim, k=1, which='SA')
```

```python
# Finding the entanglement entropy in the ground state for several cut lengths:
mfim_crit_entropies = [cu.entanglement_entropy(ground_state, x) for x in cut_lens]
```

```python
# Matching to the CFT formula:
c, cprime = cu.match_cft_entropy(L, cut_lens=cut_lens, entropies=mfim_crit_entropies)

print("Central charge:",c)
print("Renyi parameter:",cprime)
```

```python
# Plotting the exact result against the analytic result:
plt.scatter(cut_lens, mfim_crit_entropies)

xs = np.linspace(1e-10, L, 100)
plt.plot(xs, cu.cft_entropy(L, xs, c, cprime))
plt.ylim(0, .8)

plt.show()
```
